Menggunakan Library Numpy

```python
import numpy as np

# Matriks

matriksA = np.array([[15,3,1],
                    [4,7,8],
                    [3,10,6]])
matriksB = np.array([[2,4,9],
                    [6,7,8],
                    [1,7,5]])   

perkalian_skalar = matriksA*matriksB
perkalian_matriks = np.dot(matriksA,matriksB)

print("Matriks A :")
print(matriksA)

print("Matriks B :")
print(matriksB)

print("Perkalian Skalar : ")
print(perkalian_skalar)

print("Perkalian Matriks : ")
print(perkalian_matriks)
```

Perkalian Matriks Menggunakan Cara Manual (Nested Loops)

```python
# Program to multiplB two matrices using nested loops

# Matriks

A = [[15,3,1],
    [4,7,8],
    [3,10,6]]

B = [[2,4,9],
    [6,7,8],
    [1,7,5]]

# result 

result = [[0,0,0],
         [0,0,0],
         [0,0,0]]

# iterate through rows of A
for i in range(len(A)):
   # iterate through columns of B
   for j in range(len(B[0])):
       # iterate through rows of B
       for k in range(len(B)):
           result[i][j] += A[i][k] * B[k][j]

for r in result:
   print(r)
```
